```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
front = cv2.imread('front.jpg')
back = cv2.imread('back.jpg')

front = cv2.cvtColor(front, cv2.COLOR_BGR2RGB)
back = cv2.cvtColor(back, cv2.COLOR_BGR2RGB)

front = cv2.resize(front, (back.shape[1], back.shape[0]))
```

```python
fig = plt.figure(figsize=(15, 15)) 

ax = fig.add_subplot(1, 2, 1)
ax.imshow(front)
ax.set_title('front')

ax = fig.add_subplot(1, 2, 2)
ax.imshow(back)
ax.set_title('back')
```

```python
shown = (front & (256 - 16)) | (back >> 4)
plt.imshow(shown)
```

```python
hidden = (shown & 15) << 4
plt.imshow(hidden)
```
